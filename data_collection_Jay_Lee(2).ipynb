```python
key = 'SECRET'

endpoint_url = 'https://newsapi.org/v2/everything?'

query = 'Disney' # Change queries to: 'HBO' 'Disney', and run again

params = {
        'q': query,
        'pageSize': 100,
        'apiKey': key
    }


    
api_request = requests.get(endpoint_url,params)

api_request_response = json.loads(api_request.text)

articles = api_request_response['articles']
    
articles_dict = {
        'source':[],
        'author':[],
        'title':[],
        'description':[],
        'publishedAt':[],
        'content':[]
    }

for article in articles:
        
    source = article['source']['name']
    articles_dict['source'].append(source)
    print(source)

    author = article['author']
    articles_dict['author'].append(author)
    print(author)

    title = article['title']
    articles_dict['title'].append(title)
    print(title)

    description = article['description'][:200]
    articles_dict['description'].append(description)
    print(description)

    publishedAt = article['publishedAt']
    articles_dict['publishedAt'].append(publishedAt)
    print(publishedAt)

    content = article['content']
    articles_dict['content'].append(content)
    print(content)
      
    
article_dataframe = pd.DataFrame(articles_dict)

engine = create_engine("%sql mysql://USERNAME:PASSWORD@HOST/DATABASE")

article_dataframe.to_sql('news_article', engine, if_exists='append',index = False)
```
