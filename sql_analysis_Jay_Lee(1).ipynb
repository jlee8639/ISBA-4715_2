Exploratory Query 1: Non-english results may throw off the sentiment scores because the analysis tool is trained on english text. This query looks for all those non-english results by looking for letters with tildes using REGEXP

```python
%%sql 
SELECT *
FROM sentiment_analysis sa
WHERE description REGEXP '[ÁÉÍÓÚÜáéíóúü]';
```

Exploratory Query 2: This query looks for the most popular source

```python
%%sql
SELECT
    source,
    COUNT(source) AS count_of_source
FROM news_article na
GROUP BY source
ORDER BY COUNT(source) DESC;
```

Exploratory Query 3: This query looks at the sentiment scores by source to see a potential bias

```python
%%sql
SELECT
    na.source,
    COUNT(na.source) AS count_of_source,
    ROUND(SUM(sa.sentiment_score),2) AS sentiment_score,
    COUNT(CASE WHEN sentiment_category = 'positive' THEN 1 END) AS positive_count,
    COUNT(CASE WHEN sentiment_category = 'negative' THEN 1 END) AS negative_count,
    COUNT(CASE WHEN sentiment_category = 'neutral' THEN 1 END) AS neutral_count
FROM news_article na
JOIN sentiment_analysis sa 
    ON na.article_id = sa.article_id
JOIN brands b
    ON na.brand_id = b.brand_id
GROUP BY na.source
ORDER BY sentiment_score DESC;
```

Exploratory Query 4: Look for the most popular authors (authors with the most published from this dataset)

```python
%%sql
SELECT
    author,
    COUNT(author)
FROM news_article na
GROUP BY author
ORDER BY COUNT(author) DESC;
```

Exploratory Query 5: This query looks at the sentiment scores by authors to see any potential bias

```python
%%sql
SELECT
na.author,
    COUNT(na.author) AS count_of_source,
    ROUND(SUM(sa.sentiment_score),2) AS sentiment_score,
    COUNT(CASE WHEN sentiment_category = 'positive' THEN 1 END) AS positive_count,
    COUNT(CASE WHEN sentiment_category = 'negative' THEN 1 END) AS negative_count,
    COUNT(CASE WHEN sentiment_category = 'neutral' THEN 1 END) AS neutral_count
FROM news_article na
JOIN sentiment_analysis sa 
    ON na.article_id = sa.article_id
JOIN brands b
    ON na.brand_id = b.brand_id
GROUP BY na.author
ORDER BY sentiment_score DESC;
```

Question 1: How does each brand perform on sentiment scores overall?

```python
%%sql
SELECT
    brand_name,
    ROUND(SUM(sa.sentiment_score),2) AS sentiment_score
FROM news_article na
JOIN sentiment_analysis sa 
    ON na.article_id = sa.article_id
JOIN brands b
    ON na.brand_id = b.brand_id
GROUP BY b.brand_name
ORDER BY sentiment_score DESC;
```

Question 2: Are certain scores weighted more heavily than others? Look into the count of positive, negative, and neutral scores

```python
%%sql
SELECT
    brand_name,
    COUNT(CASE WHEN sentiment_category = 'positive' THEN 1 END) AS positive_count,
    COUNT(CASE WHEN sentiment_category = 'negative' THEN 1 END) AS negative_count,
    COUNT(CASE WHEN sentiment_category = 'neutral' THEN 1 END) AS neutral_count
FROM news_article na
JOIN sentiment_analysis sa 
    ON na.article_id = sa.article_id
JOIN brands b
    ON na.brand_id = b.brand_id
GROUP BY b.brand_name;
```

Question 3: Do certain popular sources produce more positive or negative articles?

```python
%%sql
WITH source_scores AS (
SELECT
    na.source,
    COUNT(na.source) AS count_of_source,
    ROUND(SUM(sa.sentiment_score),2) AS sentiment_score_total,
    COUNT(CASE WHEN sentiment_category = 'positive' THEN 1 END) AS positive_count,
    ROUND(SUM(CASE WHEN sentiment_category = 'positive' THEN sentiment_score END),2) AS positive_score_total,
    COUNT(CASE WHEN sentiment_category = 'negative' THEN 1 END) AS negative_count,
    ROUND(SUM(CASE WHEN sentiment_category = 'negative' THEN sentiment_score END),2) AS negative_score_total,
    COUNT(CASE WHEN sentiment_category = 'neutral' THEN 1 END) AS neutral_count
FROM news_article na
JOIN sentiment_analysis sa 
    ON na.article_id = sa.article_id
JOIN brands b
    ON na.brand_id = b.brand_id
GROUP BY na.source
ORDER BY sentiment_score DESC
)
SELECT *
FROM source_scores
WHERE count_of_source > 15;
```
